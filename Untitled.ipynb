###   (1m 10m)

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd

df_1 = pd.read_csv('rfdata/1m.csv', names=['rf','class'], header=None)
df_10 = pd.read_csv('rfdata/10m.csv', names=['rf','class'], header=None)
```

```python
df_concat =  pd.concat([df_1,df_10], ignore_index=True)
df_concat
```

```python
X = df_concat.iloc[:,0].values.reshape(-1, 1)
y = df_concat.iloc[:,1].values
X
y
```

###    10m 1m   .

# 1. Check Data

```python
from matplotlib import pyplot as plt
plt.scatter(X,y)
```

# 2. Split data as train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

# 3. Model

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression().fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
# Decision tree
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier().fit(X_train,y_train)
clf.score(X_test, y_test)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
df_4 = df_4.rolling(5).sum()
df_6 = df_6.rolling(5).sum()
```

```python
df_concat =  pd.concat([df_4,df_6], ignore_index=True)
df_concat = df_concat.dropna(axis = 0)
```

```python
# using pd.rolling(5).sum()
X = df_concat.iloc[:,0].values.reshape(-1, 1)
y = df_concat.iloc[:,1].values
```

```python
from matplotlib import pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression().fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
# Decision tree
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier().fit(X_train,y_train)
clf.score(X_test, y_test)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```
