# Supervised Learning
- There exists input and output sample data(training data).
- We predict the output from the input.
- Create model using the training data.


# Type of Supervised Learning

#### 1. Classification
- Predict the class label.
- ex) binary classification(yes/no).
- ex) multiples classification(tiger/lion/cat/dog...).

#### 2. Regression
- Predict the real number.
- ex) predict the yearly income based on one's education level, age, or residence.


# Generalization, Overfitting, Underfitting



source: https://www.geeksforgeeks.org/underfitting-and-overfitting-in-machine-learning/
#### 1. Generalization
- When the trained model can predict the new data exactly.

#### 2. Overfitting
- When the trained model is so complex that it fits well on the training data sets.
- However, it is hard to predict the new data.

#### 3. Underfitting
- When the trained model is so simle that it doesn't fits well on the training data sets.
- Also, it is hard to predict the new data.



source: https://www.analyticsvidhya.com/blog/2020/02/underfitting-overfitting-best-fitting-machine-learning/

#### 4. Relation between Model Complexity and Size of Dataset
- If there are various of data, the model complexity is higher
- We better to collect many data which is not overlapped and similar.
- Do not underestimate the size of dataset.
